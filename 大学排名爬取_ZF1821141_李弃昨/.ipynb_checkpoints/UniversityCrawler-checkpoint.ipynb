```python
import requests
import re
from bs4 import BeautifulSoup

def fetchUrl(url):
    '''
    Func: fetchUrl is used to get html page content
    Parameter:
    url : the url for the website 
    return : req.test as page content
    '''
    try:
        #imitate browser
        headers={'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8','User_Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.92 Safari/537.36'
}
        #get page content 
        req = requests.get(url,headers = headers)
        # check request status for throw exception
        req.raise_for_status() 
        #encoding the content based on the response from the url
        req.encoding=req.apparent_encoding 
        print('Fetch URL successful.')
        return req.text
    except requests.RequestError as e:
        print(e) #throw error
    except Exception as e:
        print(e) #throw error

def parseHtml(page,urating):
    '''
    Func: parseHtml is used to parse HTML Doc.
    Parameters：
        page: page content
        urating:container for parse result
    '''
    try:
        #parse html with html.parser,return with an beautifulsoap object
        bsoap = BeautifulSoup(page,'html.parser')
        
        # get table headers,
        #find('thead')get the whole <thead> tag,.tag func can be used multiply
        #locate the content in steps
        
        tr = bsoap.find('thead').find('tr')
        hlist = []
        # isinstance is used to judge the type of the element
        if isinstance(tr,bs4.element.Tag):
            for th in tr('th'):
                hlist.append(th.string)
            urating.append(hlist)
            
        # get table rows
        for tr in bsoap.find('tbody').children:
            blist = []
            if isinstance(tr,bs4.element.Tag):
                for td in tr('td'):
                    blist.append(td.string)
                    #print(td.string)
            urating.append(blist)
        return urating
    except Exception as e:
        print(e) 
        
def output(urating,filename):
    '''
    Func : output the result to csv file
    Parameters：
        urating:the data stored in list
        filename:the filename of the output file.
    '''
    import pandas as pd
    dataframe = pd.DataFrame(urating)
    dataframe.to_csv(filename,index=False,sep=',',header=False)
    print('CSV file : %s has been writed.'%filename)
    
def main():
    '''
    Func:调用函数
    '''
    urldict = {
                '生源质量':'http://www.zuihaodaxue.com/shengyuanzhiliangpaiming2018.html',
                '培养结果':'http://www.zuihaodaxue.com/biyeshengjiuyelv2018.html'    
    }
    indicationSystem=['生源质量',
                      '培养结果'
                     ]    
    #调用字典
    for key,url in urldict.items():
        urating = []
        page = fetchUrl(url)
        ur = parseHtml(page,urating)
        output(ur,key+'ranking2018.csv')

if __name__ == '__main__':
    main()
```

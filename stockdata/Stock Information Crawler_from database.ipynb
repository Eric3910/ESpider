```python
'''
Stock Information Crawler
example url: http://quote.eastmoney.com/sh600050.html
'''

import urllib
from bs4 import BeautifulSoup as bs
import pandas as pd
import os
import csv
import re

#Crawler website
def fetchUrl(url):
    '''
    Func: fetchUrl is used to get html page content
    Parameter:
    url : the url for the website 
    return : req.test as page content
    '''
    try:
        #imitate browser
        headers={'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8','User_Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/69.0.3497.92 Safari/537.36'
}
        #get page content 
        response=urllib.request.Request(url=url,headers=headers)
        html=urllib.request.urlopen(response)
        page=html.read().decode('gbk')
        return page
    except Exception as e:
        print(e) #throw error

#Parsing Stack Code
def getStackCode(page):
    s = r'<li><a target="_blank" href="http://quote.eastmoney.com/\S\S(.*?).html">'
    pat = re.compile(s)
    code = pat.findall(page)
    return code

#Crawler for stock information and output to csv
def output(code,filepath):

    #Get all stock code in a set
    CodeList = []
    for item in code:
        if item[0]=='6':
            CodeList.append(item)
    #get data and save in csv files
    for code in CodeList:
        print('Getting stock data of %s'%code)
        url = 'http://quotes.money.163.com/service/chddata.html?code=0'+code+\
        '&end=20181231&fields=TCLOSE;HIGH;LOW;TOPEN;LCLOSE;CHG;PCHG;TURNOVER;VOTURNOVER;VATURNOVER;TCAP;MCAP'
        print(url)
        response = urllib.request.Request(url)
        html=urllib.request.urlopen(response)
        page=html.read().decode('gbk')
        dataframe = pd.DataFrame([code,page])
        filename = filepath+code+'.csv'     
        dataframe.to_csv(filename,index=False,sep=',',header=False)
        print('CSV file : %s has been writed.'%filename)
            
def main():
    Url = 'http://quote.eastmoney.com/stocklist.html'#stock data link in Eastmoney
    filepath = '/Users/mac/Crawler/stockdata/' #Define the store directory for data file
    #Crawler
    code = getStackCode(fetchUrl(Url))
    output(code,filepath)
    
if __name__ == '__main__':
    main()
```
